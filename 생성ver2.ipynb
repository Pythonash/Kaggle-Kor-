# 데이터전처리

```python
import pandas as pd
import urllib.request
!pip install konlpy
import konlpy
from konlpy.tag import Okt
from bs4 import BeautifulSoup
from itertools import combinations as c
from google.colab import output as ot
import re
import torch
import torchsummary
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import StratifiedKFold as stk
import matplotlib.pyplot as plt
df_train = pd.read_csv('/content/train.csv')
okt = Okt()
reviews = np.array(df_train[df_train['추천 점수'] <= 2]['(추천 점수)'])
ot.clear()
stop_words = ['은','는','이','가','하','아','것','들','의','있','되','수','보','주','등','한', '듯', '단','론','다','나','에','의', '도','께','에게','으로','로', '니']
clear_sentences = []
re.compile('[^0-9ㄱ-ㅎ가-힣 ]')
for ind,i in tqdm(enumerate(reviews)):
  re_sentence = re.sub('[^0-9ㄱ-ㅎ가-힣 ]', '', i)
  re_sentence = okt.morphs(re_sentence, norm = True, stem = True)
  re_sentence = [w for w in re_sentence if not w in stop_words]
  clear_sentences.append(re_sentence)

clear_train_df_len = []
for i in clear_sentences:
  clear_train_df_len.append(len(i))
clear_train_df_len = pd.DataFrame(clear_train_df_len)

clear_train_df = np.array(clear_sentences)[clear_train_df_len[clear_train_df_len[0] <= 70].index]


word_dict = []
for review in tqdm(clear_train_df):
  for word in review:
    word_dict.append(word)
word_dict = pd.DataFrame(word_dict).drop_duplicates()


Ex_dict = np.array([
  'P'
])
Ex_dict = np.reshape(Ex_dict, (-1, 1))
word_dictionary = {}
for num, cha in enumerate(np.concatenate([Ex_dict, word_dict], axis = 0)):
  word_dictionary[cha[0]] = num

dictionary_word = {}
for num, cha in enumerate(np.concatenate([Ex_dict, word_dict], axis = 0)):
  dictionary_word[num] = cha[0]



tokenized_clear_train_df = []
for r in clear_train_df:
  tokenized = []
  for w in r:
    tokenized.append(word_dictionary[w])
  tokenized_clear_train_df.append(tokenized)
tokenized_clear_train_df = pd.DataFrame(tokenized_clear_train_df).fillna(0)
tokenized_clear_train_df

def convert_tokens_to_sentence(tokens):
  sen = []
  for token in tokens:
    k = dictionary_word[token]
    if k != 'P':
      sen.append(k)
  return sen

print(tokenized_clear_train_df.shape)
convert_tokens_to_sentence(tokenized_clear_train_df.iloc[0])
```

# 생성

```python
coding_size = 128
class Generator(torch.nn.Module):
  def __init__(self):
    super(Generator, self).__init__()

    self.MLP = torch.nn.Sequential(
        torch.nn.Linear(in_features = 128, out_features = 512),
        torch.nn.BatchNorm1d(512),
        torch.nn.ReLU(),

        torch.nn.Linear(in_features = 512, out_features = 1024),
        torch.nn.BatchNorm1d(1024),
        torch.nn.ReLU(),

        torch.nn.Linear(in_features = 1024, out_features = 32 * 70),
        torch.nn.BatchNorm1d(32 * 70),
        torch.nn.ReLU()

        # torch.nn.Linear(in_features = 2048, out_features = 4000),
        # torch.nn.BatchNorm1d(4000),
        # torch.nn.ReLU()
    )
    
    self.CTN = torch.nn.Sequential(
        torch.nn.ConvTranspose1d(32, 256, 1, 1, 0, bias = True),
        torch.nn.BatchNorm1d(256),
        torch.nn.ReLU(),

        torch.nn.ConvTranspose1d(256, 1024, 1, 1, 0, bias = True),
        torch.nn.BatchNorm1d(1024),
        torch.nn.ReLU(),

        torch.nn.ConvTranspose1d(1024, len(word_dictionary), 1, 1, 0, bias = True),
        # torch.nn.BatchNorm1d(6726),
        # torch.nn.ReLU(),

        # torch.nn.Conv1d(2048, 2794, 1, 1, 0, bias = True),
        # torch.nn.Flatten(),
        # torch.nn.ReLU()
        )

  def forward(self,x):
    x = self.MLP(x)
    x = x.view(-1, 32, 70)
    x = self.CTN(x)
    return x

class Discriminator(torch.nn.Module):
  '''
  token_num: 음절사전 토큰 개수
  embedding_dim: 몇 차원으로 찢을지
  padding_dix: 패딩 토큰 지정
  '''
  def __init__(
      self, 
      token_num,
      embedding_dim,
      padding_idx = 0,
      *args,
      **kwargs
               ):
    super(Discriminator, self).__init__()
    self.token_num = token_num + 1
    self.embedding_dim = embedding_dim
    self.padding_idx = padding_idx

 # (Batch, input_features) -> (Batch, input_features, embedding_size) -> view로 차원 조절해주고 -> (Batch, embedding_size, input_features)
    self.sentence_embedding = torch.nn.Embedding(
            num_embeddings = self.token_num,
            embedding_dim = self.embedding_dim,
            padding_idx = self.padding_idx
            )
    
    self.sentence_encoder_block1_cnn1 = torch.nn.Conv1d(in_channels = self.embedding_dim, out_channels = 16, kernel_size = 1, stride =1, padding = 0)
    self.sentence_encoder_block1_bn1 = torch.nn.BatchNorm1d(num_features = 16)
    self.sentence_encoder_block1_act1 = torch.nn.ReLU()

    self.pooling0 = torch.nn.MaxPool1d(2,2)

    self.sentence_encoder_block1_cnn2 = torch.nn.Conv1d(in_channels = 16, out_channels = 32, kernel_size = 3, stride =2, padding = 0)
    self.sentence_encoder_block1_bn2 = torch.nn.BatchNorm1d(num_features = 32)
    self.sentence_encoder_block1_act2 = torch.nn.ReLU()

    self.pooling1 = torch.nn.MaxPool1d(2,2)

    self.sentence_flatten = torch.nn.Flatten()
    self.sentence_linear1 = torch.nn.Linear(256, out_features = 128)
    self.sentence_bn1 = torch.nn.BatchNorm1d(num_features = 128)
    self.sentence_act1 = torch.nn.LeakyReLU()
    self.dropout1 = torch.nn.Dropout(0.2)

    self.sentence_linear2 = torch.nn.Linear(in_features = 128, out_features = 64)
    self.sentence_bn2 = torch.nn.BatchNorm1d(num_features = 64)
    self.sentence_act2 = torch.nn.LeakyReLU()

    self.linear = torch.nn.Linear(in_features = 64, out_features = 1)
    self.act = torch.nn.Sigmoid()

  def forward(self, x):
    x = self.sentence_embedding(x)
    channel, length = x.shape[2], x.shape[1]
    x = x.view(-1, channel, length)
    x = self.sentence_encoder_block1_cnn1(x)
    x = self.sentence_encoder_block1_bn1(x)
    x = self.sentence_encoder_block1_act1(x)
    x = self.pooling0(x)
    x = self.sentence_encoder_block1_cnn2(x)
    x = self.sentence_encoder_block1_bn2(x)
    x = self.sentence_encoder_block1_act2(x)
    x = self.pooling1(x)

    x = self.sentence_flatten(x)
    x = self.sentence_linear1(x)
    x = self.sentence_bn1(x)
    x = self.sentence_act1(x)
    x = self.dropout1(x)
    x = self.sentence_linear2(x)
    x = self.sentence_bn2(x)
    x = self.sentence_act2(x)

    logit = self.linear(x)
    logit = self.act(logit)
    return logit

print(Discriminator(len(word_dictionary) + 1, 128)(torch.randint(low =0, high = len(word_dictionary), size = (2, 70))))
Generator()(torch.randn((2, coding_size))).clone().detach().numpy().shape

Real_text_loader = torch.utils.data.DataLoader(torch.tensor(np.array(tokenized_clear_train_df), dtype = torch.int64, device = 'cuda'), shuffle = True, batch_size = 32, drop_last = True)
gan_loss_fn = torch.nn.BCELoss()
generator = Generator()
generator.to(device='cuda')
discriminator = Discriminator(len(word_dictionary) + 1, 128)
discriminator.to(device = 'cuda')
generator_optimizer = torch.optim.SGD (generator.parameters(), lr = 0.00002, momentum = 0.9)
discriminator_optimizer = torch.optim.Adam(discriminator.parameters(), lr = 0.003)

for epoch in range(10):
  print(f'''Epoch {epoch + 1} 시작''')
  
  for batch_text in Real_text_loader:
    gen_input1 = torch.randn((32, 128), dtype = torch.float32)
    gen_input1 = gen_input1.to(device = 'cuda')
    fake1 = torch.max(generator(gen_input1).view(-1, 70, len(word_dictionary)), axis = 2)[1]
    fake1 = torch.tensor(fake1, dtype = torch.int64, device = 'cuda')

    X_fake_real = torch.concat([fake1, batch_text], axis = 0)
    y1_1 = torch.zeros((32,1), dtype = torch.float32, device = 'cuda')
    y1_2 = torch.ones((32, 1), dtype = torch.float32, device = 'cuda')
    y1 = torch.concat([y1_1, y1_2], axis = 0)

    discriminator.train()
    loss_d = gan_loss_fn(discriminator(X_fake_real), y1)
    discriminator_optimizer.zero_grad()
    loss_d.backward()
    discriminator_optimizer.step()
    discriminator.eval()

    y2 = torch.ones((32, 1), dtype = torch.float32, device = 'cuda')
    generator.train()
    fake2 = torch.max(generator(gen_input1).view(-1, 70, len(word_dictionary)), axis = 2)[1]
    fake2 = torch.tensor(fake2, dtype = torch.int64, device = 'cuda')
    loss_g = gan_loss_fn(discriminator(fake2), y2)
    generator_optimizer.zero_grad()
    loss_g.backward()
    generator_optimizer.step()
    generator.eval()
    print(loss_d)
    print(loss_g)
    print(' '.join(convert_tokens_to_sentence(fake2.to(device = 'cpu').clone().detach().numpy()[16])))
```

# 사전학습

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt", filename="ratings_train.txt")
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt", filename="ratings_test.txt")
train_data = pd.read_table('ratings_train.txt')
test_data = pd.read_table('ratings_test.txt')
pre_reviews = pd.concat([train_data, test_data],axis = 0).dropna().drop(['id'], axis = 1).reset_index(drop=True)

# 제대로 할 때에는 여기부분 주석
pre_reviews_0 = pre_reviews[pre_reviews['label'] == 0].reset_index(drop = True)
pre_reviews_1 = pre_reviews[pre_reviews['label'] == 1].reset_index(drop = True)

index_0 = np.random.randint(low = 0, high = len(pre_reviews_0), size = (5000))
index_1 = np.random.randint(low = 0, high = len(pre_reviews_1), size = (5000))

pre_reviews_0 = pre_reviews_0.iloc[index_0]
pre_reviews_1 = pre_reviews_1.iloc[index_1]

pre_reviews = pd.concat([pre_reviews_0, pre_reviews_1],axis = 0).reset_index(drop=True).sample(frac = 1)
# 여기까지

pre_clear_sentences = []
for ind,i in tqdm(enumerate(pre_reviews['document'])):
  re_sentence = re.sub('[^0-9ㄱ-ㅎ가-힣 ]', '', i)
  re_sentence = okt.morphs(re_sentence, norm = True, stem = True)
  re_sentence = [w for w in re_sentence if not w in stop_words]
  pre_clear_sentences.append(re_sentence)

pre_clear_train_df_len = []
for i in pre_clear_sentences:
  pre_clear_train_df_len.append(len(i))
pre_clear_train_df_len = pd.DataFrame(pre_clear_train_df_len)

pre_clear_train_df = np.array(pre_clear_sentences)[pre_clear_train_df_len[pre_clear_train_df_len[0] <= 70].index]


pre_word_dict = []
for review in tqdm(pre_clear_train_df):
  for word in review:
    pre_word_dict.append(word)
pre_word_dict = pd.DataFrame(pre_word_dict).drop_duplicates()


pre_word_dictionary = {}
for num, cha in enumerate(np.concatenate([Ex_dict, pre_word_dict], axis = 0)):
  pre_word_dictionary[cha[0]] = num

pre_dictionary_word = {}
for num, cha in enumerate(np.concatenate([Ex_dict, pre_word_dict], axis = 0)):
  pre_dictionary_word[num] = cha[0]



pre_tokenized_clear_train_df = []
for r in pre_clear_train_df:
  pre_tokenized = []
  for w in r:
    pre_tokenized.append(pre_word_dictionary[w])
  pre_tokenized_clear_train_df.append(pre_tokenized)
pre_tokenized_clear_train_df = pd.DataFrame(pre_tokenized_clear_train_df).fillna(0)


def pre_convert_tokens_to_sentence(tokens):
  sen = []
  for token in tokens:
    k = pre_dictionary_word[token]
    if k != 'P':
      sen.append(k)
  return sen

def length_set(df):
  if df.shape[1] < 70:
    short = 70 - df.shape[1]
  pad_set = pd.DataFrame(np.zeros(len(df)))
  for i in range(short):
    df = pd.concat([df, pad_set], axis = 1)
  df = df.T.reset_index(drop = True).T
  return df


pre_tokenized_clear_train_df = length_set(pre_tokenized_clear_train_df)

print(pre_tokenized_clear_train_df.shape)
pre_convert_tokens_to_sentence(pre_tokenized_clear_train_df.iloc[0])



```

```python
coding_size = 128
class Pre_Generator(torch.nn.Module):
  def __init__(self):
    super(Pre_Generator, self).__init__()
    
    self.Embedding = torch.nn.Embedding(
            num_embeddings = len(pre_word_dictionary) + 1,
            embedding_dim = 128,
            padding_idx = 0
            )
    
    self.MLP = torch.nn.Sequential(
        torch.nn.Linear(in_features = 128, out_features = 512),
        torch.nn.BatchNorm1d(512),
        torch.nn.ReLU(),

        torch.nn.Linear(in_features = 512, out_features = 1024),
        torch.nn.BatchNorm1d(1024),
        torch.nn.ReLU(),

        torch.nn.Linear(in_features = 1024, out_features = 32 * 70),
        torch.nn.BatchNorm1d(32 * 70),
        torch.nn.ReLU()

        # torch.nn.Linear(in_features = 2048, out_features = 4000),
        # torch.nn.BatchNorm1d(4000),
        # torch.nn.ReLU()
    )
    
    self.CTN = torch.nn.Sequential(
        torch.nn.ConvTranspose1d(32, 256, 1, 1, 0, bias = True),
        torch.nn.BatchNorm1d(256),
        torch.nn.ReLU(),

        torch.nn.ConvTranspose1d(256, 1024, 1, 1, 0, bias = True),
        torch.nn.BatchNorm1d(1024),
        torch.nn.ReLU(),

        torch.nn.ConvTranspose1d(1024, len(pre_word_dictionary), 1, 1, 0, bias = True),
        # torch.nn.BatchNorm1d(6726),
        # torch.nn.ReLU(),

        # torch.nn.Conv1d(2048, 2794, 1, 1, 0, bias = True),
        # torch.nn.Flatten(),
        # torch.nn.ReLU()
        )

  def forward(self,x):
    x = self.Embedding(x)
    x = torch.mean(x, axis = 1)
    x = self.MLP(x)
    x = x.view(-1, 32, 70)
    x = self.CTN(x)
    return x


# Pre_Generator()(torch.randint(low = 0, high = len(word_dictionary), size = (2, 200)))

pre_Real_text_loader = torch.utils.data.DataLoader(torch.tensor(np.array(pre_tokenized_clear_train_df), dtype = torch.float32, device = 'cuda'), shuffle = True, batch_size = 32, drop_last = True)


loss_fn = torch.nn.MSELoss()
pre_generator = Pre_Generator()
pre_generator.to(device='cuda')
pre_generator_optimizer = torch.optim.Adam(pre_generator.parameters(), lr = 0.003)

for epoch in range(10):
  print(f'''Epoch {epoch + 1} 시작''')
  
  for batch_text in pre_Real_text_loader:
    y = torch.max(pre_generator(torch.tensor(batch_text, dtype = torch.int64, device = 'cuda')).view(-1, 70, len(pre_word_dictionary)), axis = 2)[1]
    y = torch.tensor(y, dtype = torch.float32, device = 'cuda')

    pre_generator.train()
    loss_ = loss_fn(y, batch_text)
    pre_generator_optimizer.zero_grad()
    loss_.backward()
    pre_generator_optimizer.step()
    pre_generator.eval()
    print(loss_)
    print(' '.join(pre_convert_tokens_to_sentence(y.to(device = 'cpu').clone().detach().numpy()[16])))
```
